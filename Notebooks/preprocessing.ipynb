```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../Data/heart.csv")
df.head(10)
```

```python
#droppig columns that have low correlation score in eda
df_processed = df.drop(['Cholesterol', 'MaxHR'], axis=1)
df_processed.head(10)
```

```python
encoder = OneHotEncoder(sparse_output=False)
categorical_columns = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]
one_hot_encoded = encoder.fit_transform(df[categorical_columns])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
df_processed_encoded = pd.concat([df.drop(categorical_columns, axis=1), one_hot_df], axis=1)
df_processed_encoded.head(10)
```

Preprocessing decision:
- Dropped columns: "Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"
- Categorical encoding: OneHot
- Scaling: StandardScaler
